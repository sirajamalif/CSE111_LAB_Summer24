```python
class Student:
    def __init__(self, name, id, dept='CSE'):
        self.name = name
        self.id = id
        self.dept = dept
        self.sugg = ""

    def dailyEffort(self, effort=0):
        self.effort = effort
        if self.effort <= 2:
            self.sugg = "Should give more effort!"
        elif self.effort <= 4:
            self.sugg = "Keep up the good work!"
        else:
            self.sugg = "Excellent! Now motivate others."

    def printDetails(self):
        print(f"Name: {self.name}\nID: {self.id}\nDepartment: {self.dept}\nDaily Effort: {self.effort} hour(s)\nSuggestion: {self.sugg}")

#Driver code:
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
class Farmer:
    def __init__(self, name=None):
        self.name = name
        if isinstance(self.name, int):
            print(f"Welcome to your farm. Your farm ID is {self.name}!")
        elif isinstance(self.name, str):
            print(f"Welcome to your farm, {self.name}!")
        else:
            print("Welcome to your farm!")
        self.crops = []
        self.fishes = []

    def addCrops(self, *crops):
        if not crops:
            print("No crop added.")
        else:
            for crop in crops:
                self.crops.append(crop)
            print(f"{len(crops)} crop(s) added.")

    def addFishes(self, *fishes):
        if not fishes:
            print("No fish added.")
        else:
          for fish in fishes:
            self.fishes.append(fish)
          print(f"{len(fishes)} fish(s) added.")

    def showGoods(self):
        if not self.crops and not self.fishes:
            print("You don't have any crop(s) or fish(s).")
        elif self.crops and not self.fishes:
            print(f"You have {len(self.crops)} crop(s): \n{', '.join(self.crops)} \nYou don't have any fish(s).")
        elif not self.crops and self.fishes:
            print(f"You don't have any crop(s). \nYou have {len(self.fishes)} fish(s): \n{', '.join(self.fishes)}")
        else:
            print(f"You have {len(self.crops)} crop(s): \n{', '.join(self.crops)} \nYou have {len(self.fishes)} fish(s): \n{', '.join(self.fishes)}")

# Driver Code
f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
class Sphere:
    def __init__(self, id, radius=1, color="White"):
        self.id = id
        self.radius = radius
        self.color = color

    def printDetails(self):
        print(f"Sphere ID: {self.id}")
        print(f"Color: {self.color}")
        print(f"Volume: {self.calculate_volume()}")

    def calculate_volume(self):
        return (4 / 3) * 3.1416 * self.radius**3

    def merge_sphere(self, *spheres):
        total_volume = self.calculate_volume()
        for sphere in spheres:
            total_volume += sphere.calculate_volume()
            self.radius = (total_volume / ((4 / 3) * 3.1416))**(1 / 3)
            if self.color != sphere.color:
              self.color = "Mixed Color"
        print("Spheres are being merged")

#Write your code here
sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```

```python
class Foodie:
    def __init__(self, name):
        self.name = name
        self.menu = menu
        self.foods_quants = {}
        self.total = 0

    def order(self, *items):
        for item in items:
            food, quant = item.split('-')
            quant = int(quant)
            if food in self.menu:
                individual_price = self.menu[food] * quant
                if food in self.foods_quants:
                    self.foods_quants[food] += quant
                else:
                    self.foods_quants[food] = quant
                self.total += individual_price
                print(f"Ordered - {food}, quantity - {quant}, price (per Unit) - {self.menu[food]}.")
                print(f"Total price - {individual_price}")
        self.foods = list(self.foods_quants.keys())

    def pay_tips(self, tips=0):
        self.tips = tips
        if not self.tips:
            print("No tips to the waiter.")
        else:
            self.total += self.tips
            print(f"Gives {self.tips}/- tips to the waiter.")

    def show_orders(self):
        if not self.foods_quants:
            print(f"{self.name} has 0 item(s) in the cart.")
            print("Items: []")
        else:
            print(f"{self.name} has {len(self.foods)} item(s) in the cart.")
            print(f"Items: {self.foods}")
        print(f"Total spent: {self.total}.")
        return ""


# Driver Code:
menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}

f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
class Department:
    def __init__(self,dept="CHE Department",sec=5):
        self.department=dept
        self.section=sec
        self.avg=0
        print(f"The {self.department} has {self.section} sections")

    def add_students(self, *students):
        section_count = 0
        total_students = 0

        for student_count in students:
            section_count += 1
            total_students += student_count

        if section_count == self.section:
            self.avg = round(total_students / self.section, 2)
            print(f"The {self.department} has an average of {self.avg} students in each section")
        else:
            print(f"The {self.department} doesn't have {section_count} sections")


    def merge_Department(self, *loc):
        total=self.avg*self.section
        for cl in loc:
            total+=(cl.avg*cl.section)
            output=f"The {cl.department} is merged to {self.department}.\n"
        self.avg=round(total/self.section,2)
        output+=f"Now the {self.department} has an average of {self.avg} students in each section."
        return output
#Driver Code
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21, 30, 40, 36, 10, 32, 27, 51, 45, 15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Shopidify:
    def __init__(self, username=None):
        self.username = username
        if isinstance(self.username,str):
          print(f"Welcome {self.username} to Shopidify")
        else:
          print("Welcome to Shopidify")
        self.cart = {}
        self.purchase_history = []

    def add_to_cart(self, item, quantity=1):
        if isinstance(item, list):
            for i in range(0, len(item), 2):
                self.add_to_cart(item[i], item[i+1])
        else:
            if item in self.cart:
                self.cart[item] += quantity
            else:
                self.cart[item] = quantity

    def display_cart(self):
        print("Items in the cart for ", end="")
        if self.username:
            print(f"{self.username}:", end="")
        else:
            print("Guest:", end="")

        for item in self.cart:
            print(f"\n- {item}: {self.cart[item]}x", end="")



    def checkout(self):
        history_copy = {}
        for item in self.cart:
            history_copy[item] = self.cart[item]
        self.purchase_history.append(history_copy)
        self.cart = {}
        if self.username:
            print("Checkout completed for", self.username)
        else:
            print("Checkout completed for Guest")

    def display_history(self):
      print("Purchase history for ", end="")
      if self.username:
          print(f"{self.username}:", end="")
      else:
          print("Guest:", end="")

      for i in range(len(self.purchase_history)):
          print(f"\nTransaction {i + 1}:", end="")
          for item in self.purchase_history[i]:
              print(f"\n- {item}: {self.purchase_history[i][item]}x", end="")



# Driver Code
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
class Author:
    def __init__(self, name=""):
        self.name = name
        self.books = {}

    def addBook(self, title, genre):
        if self.name == "":
            print("A book cannot be added without an author name")
        else:
            if genre not in self.books:
                self.books[genre] = [title]
            else:
                self.books[genre].append(title)

    def setName(self, name):
        self.name = name

    def printDetail(self):
        total_books = 0
        for books in self.books.values():
          total_books += len(books)
        print(f"Number of Book(s): {total_books}")
        print(f"Author Name: {self.name}")

        for genre, titles in self.books.items():
            print(f"{genre}: {', '.join(titles)}")



a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")

```
