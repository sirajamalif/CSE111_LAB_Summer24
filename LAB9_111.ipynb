```python
class NikeBangladesh:
  branches_opened = []
  product_dict={'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
  sold = 0
  def __init__(self,outlet):
    self.outlet = outlet
    self.outlet_dict = {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
    self.outlet_sold = 0
    NikeBangladesh.branches_opened.append(outlet)

  def restockProducts(self,new_dict):
    for key,value in new_dict.items():
      if key in NikeBangladesh.product_dict and key in self.outlet_dict:
        NikeBangladesh.product_dict[key] += value
        self.outlet_dict[key] += value
      else:
        pass

  def productSold(self,sold_dict):
    for key,value in sold_dict.items():
      if key in NikeBangladesh.product_dict and key in self.outlet_dict:
        NikeBangladesh.product_dict[key] -= value
        self.outlet_dict[key] -= value
        NikeBangladesh.sold += value
        self.outlet_sold += value
      else:
        pass

  def details(self):
    print(f"Nike {self.outlet} outlet:")
    print("Products Currently Stocked:")
    print(self.outlet_dict)
    print(f"Sold: {self.outlet_sold}")


  @classmethod
  def status(cls):
    print("Nike Bangladesh Status:")
    print(f"Branches Opened:  {cls.branches_opened}")
    print("Currently Stocked")
    print(cls.product_dict)
    print(f"Sold: {cls.sold}")


print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
class Player:
    total = 0
    total_players = []

    def __init__(self, player=None, jnum=10, country=None):
        self.player = player
        if isinstance(self.player,str):
          Player.total_players.append(self.player)
        else:
          pass
        self.jnum = jnum
        self.country = country
        Player.total += 1

    def set_name(self, name):
        self.player = name
        Player.total_players.append(self.player)

    def set_team(self, team):
        self.country = team

    def set_number(self, number):
        self.jnum = number

    def player_detail(self):
        return f"Player Name: {self.player}\nJersey Number: {self.jnum}\nCountry: {self.country}"

    @classmethod
    def info(cls):
        print("Total number of players:", cls.total)
        cls.str_players = ', '.join(cls.total_players)
        print("Players enlisted so far:", cls.str_players)

print("Total number of players:", Player.total)
print("---------------------------")
p1 = Player()
p1.set_name("Neymar")
p1.set_team("Brazil")
print(p1.player_detail())
print('========================')
Player.info()
print("---------------------------")
p2 = Player("Ronaldo")
p2.set_number(7)
p2.set_team("Portugal")
print(p2.player_detail())
print('========================')
Player.info()
print("---------------------------")
p3 = Player("Messi")
p3.set_team("Argentina")
print(p3.player_detail())
print('========================')
Player.info()
print("---------------------------")
p4 = Player("Mbappe", 10, "France")
print(p4.player_detail())
print('========================')
Player.info()
```

```python
class SultansDine:
    total_branches = 0
    total_sell = 0
    total_list = []

    def __init__(self, branch):
        self.branch = branch
        SultansDine.total_branches += 1
        self.branch_sell = 0

    def sellQuantity(self, quant):
        if quant < 10:
            self.branch_sell = quant * 300
        elif quant < 20:
            self.branch_sell = quant * 350
        else:
            self.branch_sell = quant * 400

        SultansDine.total_sell += self.branch_sell
        SultansDine.total_list.append((self.branch, self.branch_sell))

    def branchInformation(self):
        print(f"Branch Name: {self.branch}")
        print(f"Branch Sell: {self.branch_sell} Taka")

    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.total_branches}")
        print(f"Total Sell: {cls.total_sell}")

        for branch, sell in cls.total_list:
            sell_percentage = (sell / cls.total_sell) * 100
            print(f"Branch Name: {branch}, Branch Sell: {sell} Taka")
            print(f"Branch consists of total sell's: {sell_percentage:.2f}%")





SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
