```python
class Triangle:
  def __init__(self,base,height):
    self.__base = base
    self.__height = height
  def getBase(self):
    return self.__base
  def getHeight(self):
    return self.__height
  def area(self):
    return .5*self.__base*self.__height
  def setBase(self,base):
    self.__base = base
  def setHeight(self,height):
    self.__height = height

t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())


t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
class Spaceship:
    def __init__(self,name, capacity) :
        self.name = name
        self.capacity = capacity
        self.count = 0
        self.cargolist = []

    def load_cargo(self,obj):
        self.count += obj.get_cargo_weight()
        if self.count > self.capacity :
            print(f"Warning: Unable to load {obj.get_cargo_name()} inside {self.name}. Exceeds capacity by {self.count - self.capacity }.")
        else:
          pass
        self.cargolist.append(obj.get_cargo_name())

    def display_details(self):
        print(f"Spaceship Name: {self.name}")
        print(f"Capacity: {self.capacity}")
        print(f"Current Cargo Weight: {self.count}")
        print(f"Cargo: {self.cargolist}")


class Cargo:

    def __init__(self,name, weight) :
        self.__name = name
        self.__weight = weight

    def get_cargo_name(self):
        return  self.__name

    def get_cargo_weight(self):
        return  self.__weight


# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
class BracuStudent:
    def __init__(self,name,home):
        self.name = name
        self.home = home
        self.hasBusPass = False

    def get_pass(self):
        self.hasBusPass = True

    def show_details(self):
        print(f"Student Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have Bus Pass? {self.hasBusPass}")


class BracuBus:
    def __init__(self,route,max_pass=2):
        self.route = route
        self.max_pass = max_pass
        self.pass_list = []
        self.count = 0

    def board(self, *objs):
        if not objs:
          print("No passengers!")
        else:
          for obj in objs:
            if obj.hasBusPass == False:
              print("You don't have a bus pass!")
            elif obj.home != self.route:
              print("You got on the wrong bus!")
            else:
              if self.count < self.max_pass:
                self.count += 1
                self.pass_list.append(obj.name)
                print(f"{obj.name} boarded the bus.")
              elif self.count == self.max_pass:
                print("Bus is full!")



    def show_details(self):
        print(f"Bus Route: {self.route}")
        print(f"Passengers Count: {self.count} (Max: {self.max_pass})")
        print(f"Passengers on Board: {self.pass_list}")


st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
class Library:
  def __init__(self, name, book_info):
    self.name = name
    self.book_info = book_info
    self.borrower_info = {}

  def details(self):
    print(f"{self.name} Library Details")
    print(f"Borrower Details: \n{self.borrower_info}")
    print(f"Books availability: \n{self.book_info}")

  def add_reader(self, reader):
    self.borrower_info[reader.name] = reader.total_books

class Reader:
  def __init__(self, name):
    self.name = name
    self.total_books = 0
    self.book_info = {}

  def borrow(self, library, *books):

    add_reader = False

    for book in books:
      if self.total_books >= 5:
        print("You cannot borrrow more than 5 books")
        break
      else:
        add_reader = True
        if library.book_info[book] <= 0:
          print(f"{book} books are not available at the moment")
        else:
          self.total_books += 1
          library.book_info[book] -= 1
          if book in self.book_info:
            self.book_info[book] += 1
          else:
            self.book_info[book] = 1
          print(f"{book} book is borrowed successfully")
      if add_reader == True:
        library.add_reader(self)

  def readerInfo(self, book_type = "all"):

    if book_type == "all":
      print(f"{self.name}, you have {self.total_books} book(s) with you")
      for book in self.book_info:
        print(f"Books on {book}: {self.book_info[book]}")
    else:
      print(f"{self.name}, you have {self.book_info[book_type]} {book_type} book(s) with you")


L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
